## OUTLINE:
- cursor.fetchall() vs cursor.fetch()

```python
import sqlite3

#create a connect object:
conn = sqlite3.connect('movies.db')

#create a cursor object:
cursor = conn.cursor()
```

```python
cursor.execute(
    ''' SELECT name
    FROM sqlite_master
    WHERE type = 'table'
    ''').fetchall() #to fetch the result of the SQL query's execution
```

> Let's work with the join_table.

```python
cursor.execute('''
SELECT *
FROM join_table''').fetchall()
```

- The `.fetchall()` method extracts data from the table and returns output as a string.
- Since the `.fetchall()` method only returns the values within the rows, not the name of the columns. Hence, we need to use the sqlite-master table to extract the columns' names.

```python
cursor.execute('''SELECT *
FROM sqlite_master
WHERE type = 'table'
AND name = 'join_table' 
''').fetchall()
```

```python
execute = cursor.execute('''
SELECT *
FROM join_table''')

movies = []
names = []
years = []
ratings1 = []

for row in execute:
    movie_id, name, year, rating = row
    movies.append(movie_id)
    names.append(name)
    years.append(year)
    ratings1.append(rating)

table = {'MOVIE_ID':movies, 
         'NAME': names,
         'YEARS': years,
         'RATING': ratings1}
print(table)
    
```

```python
#tunr the dictionary into a pd Df:
import pandas as pd

ratings_table = pd.DataFrame( table, columns = table.keys())
ratings_table.head()
```

```python
#save the table into a csv file:
ratings_table.to_csv('movie_ratings.csv', #file_name
                     encoding = 'utf-8', #encoding
                     index = False)  #index = True, there will be an index column
```
